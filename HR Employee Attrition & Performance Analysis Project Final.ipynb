## 📊 8. Attrition Rate Summary
A consolidated summary of attrition rates across key categories
to provide actionable insights for HR decision making.

```python
# Attrition Rate Summary Table
categories = ['Department', 'CF_age band', 'Job Role', 'Over Time', 'Marital Status']

for col in categories:
    print(f"\n=== Attrition Rate by {col} ===")
    summary = df.groupby(col)['Attrition'].value_counts(normalize=True).mul(100).round(1).unstack()
    print(summary)
    print("-"*50)
```

## 📌 Key Findings & Insights

### 🔴 Attrition Overview
- Overall attrition rate is **16.1%** — roughly 1 in 6 employees
  left the organization.

### 🏢 Attrition by Department
- **Sales** has the highest attrition **(20.6%)** followed by
  **HR (19%)** and **R&D (13.8%)**
- R&D department retains employees the best

### 👶 Attrition by Age Band
- **Under 25** employees have the highest attrition **(39.2%)** —
  nearly 4 in 10 young employees leave
- **35-44** age group is the most stable with only **10.1%** attrition
- Young employees are the highest attrition risk group

### 💼 Attrition by Job Role
- **Sales Representatives (39.8%)** and **Laboratory Technicians (23.9%)**
  have the highest attrition
- **Research Directors (2.5%)** and **Managers (4.9%)** are the
  most stable roles
- Lower paid roles show significantly higher attrition

### ⏰ Attrition by Overtime
- Employees working overtime have **3x higher attrition (30.5%)**
  compared to those who don't **(10.4%)**
- Overtime is the **single strongest attrition driver**

### 💍 Attrition by Marital Status
- **Single employees (25.5%)** are most likely to leave
- **Divorced (10.1%)** and **Married (12.5%)** employees are
  more stable

### 🔥 Correlation Insights
- **Monthly Income & Job Level (0.95):** Career growth is the
  biggest driver of compensation
- **Performance Rating & Salary Hike (0.77):** Better performers
  are rewarded with higher salary hikes
- **Total Working Years & Job Level (0.78):** Experience directly
  translates to career progression
- **Years At Company & Years With Manager (0.77):** Long-term
  employees build strong manager relationships

### 💰 Compensation Gap
- Managers earn **6.5x more (₹17,182)** than Sales
  Representatives **(₹2,626)**
- Low compensation in junior roles is a major attrition risk

### 🏆 Performance
- Performance ratings are consistent across all departments:
  **R&D (3.16)**, **HR (3.14)**, **Sales (3.14)**
- Training frequency is uniform across roles
  **(avg 2.6 - 3.0 sessions/year)**

---

## ✅ Conclusion
This analysis of **1,470 employees** reveals critical attrition
patterns across the organization. **Sales Representatives,
Laboratory Technicians, and employees under 25** are at the
highest risk of leaving. **Overtime is the single biggest
attrition driver** — employees working overtime are 3x more
likely to leave. A stark **6.5x compensation gap** between
senior and junior roles further fuels dissatisfaction.

### 🎯 Recommendations for HR Teams:
- 📉 **Reduce overtime** especially in Sales and R&D departments
- 💰 **Revisit compensation** for junior roles like Sales
  Representatives and Lab Technicians
- 🌱 **Create career growth paths** for young employees under 25
- 🏆 **Recognize high performers** with better incentives and
  salary hikes
- 👥 **Focus retention efforts** on single employees and
  those in high-attrition roles
